```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the dataset
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
df = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

# Check the first few rows of the dataset
print(df.head())
```

```python
# Plot the data
plt.figure(figsize=(10,6))
plt.plot(df.index, df['Passengers'])
plt.title('AirPassengers Time Series')
plt.xlabel('Time')
plt.ylabel('Number of Passengers')
plt.show()

```

```python
# Decompose the time series
decomposition = seasonal_decompose(df['Passengers'])

# Plot the original data, the trend, the seasonality, and the residuals
plt.figure(figsize=(12,8))
plt.subplot(411)
plt.plot(df['Passengers'], label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(decomposition.trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(decomposition.seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(decomposition.resid, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform Dickey-Fuller test:
print('Results of Dickey-Fuller Test:')
dftest = adfuller(df['Passengers'], autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print(dfoutput)
```

```python
# Create a differenced series
diff = df.diff().dropna()

# Perform Dickey-Fuller test on the differenced series
print('Results of Dickey-Fuller Test on Differenced Series:')
dftest = adfuller(diff['Passengers'], autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print(dfoutput)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Create figure
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(12,8))

# Plot the ACF of the differenced series
plot_acf(diff, ax=ax1)

# Plot the PACF of the differenced series
plot_pacf(diff, ax=ax2)

plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA

# Fit the ARIMA model
model = ARIMA(df['Passengers'], order=(2,1,2))
model_fit = model.fit(disp=0)

# Summary of the model
print(model_fit.summary())
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model
model = ARIMA(df['Passengers'], order=(2,1,2))
model_fit = model.fit()

# Summary of the model
print(model_fit.summary())
```

```python
# Plot residual errors
residuals = pd.DataFrame(model_fit.resid)
fig, ax = plt.subplots(1,2)
residuals.plot(title="Residuals", ax=ax[0])
residuals.plot(kind='kde', title='Density', ax=ax[1])
plt.show()
```

```python
# Forecast
n_steps = 12
forecast = model_fit.forecast(steps=n_steps)

# Create a new dataframe to hold the forecast
forecast_index = pd.date_range(start='1961-01', periods=n_steps, freq='MS')
forecast_df = pd.DataFrame(forecast, index=forecast_index, columns=['Forecast'])

# Plot the original data and the forecast
plt.figure(figsize=(12, 8))
plt.plot(df.index, df['Passengers'], label='Original')
plt.plot(forecast_df.index, forecast_df['Forecast'], color='red', label='Forecast')
plt.title('AirPassengers Time Series - Original and Forecast')
plt.xlabel('Time')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```
